```python
# QUESTION ONE.

#Take input for n and text
n=int(input ("please input value of n: "))
textInconsoleAsstring = input("enter text")

# using if and else statemet
#if n is divisible by both 3 and 5
if n% 3 == 0 and n% 5 ==0 :
    #print blackpink  
    text_to_print="BlackPink"
    print (text_to_print)
    
# else n is divisible by 3
elif n% 3 == 0:
    #print Black text
    text_to_print = "Black"
    print(text_to_print)
    
#else is divisible by 5
elif n% 5 == 0:
    #print pink
    text_to_print = "Pink"
    print (text_to_print)
    
# else n is divisible by 2
elif n% 2 == 0:
    #print input text backwards
    text_to_print= textInconsoleAsstring [::-1]

#%%
# question two

from matplotlib import pyplot as plt
#creating empty list

myNumbers=[]

def in_float(element):
    '''the function will retrun true if element is
       possible to convert float else return false ok'''
    
    try:
        float(element)
        return True
    except ValueError:
        return False
element=''
while True:
    element=input("Enter a number or * for stop : ")
    if element=="*":
        break
    return_value=in_float(element)
    if return_value==True:
        myNumbers.append(float(element))
    if return_value==False:
        print("could not convert string to float")
print("Given numbers are : ",myNumbers)        

#finding mean
summ=0.0
for number in myNumbers:
    summ+=number
print("The mean of the given numbers is", summ/len(myNumbers))    
        
#finding median
no = len(myNumbers)
myNumbers.sort()
if no % 2 == 0:
    median1 = myNumbers[no//2]
    median2 = myNumbers[no//2 - 1]
    median = (median1 + median2)/2
else:
    median = myNumbers[no//2]
print("The median of the given numbers is", str(median))


#finding mean
mean=summ/len(myNumbers)
myNumber=list(map(int,myNumbers))
variance = sum([((x - mean) ** 2) for x in myNumber]) / len(myNumber)
res = variance ** 0.5
# Printing result
print("Standard deviation of sample is : " + str(res))

#ploting histogram
plt.hist(myNumbers)
plt.xlabel("X-axies")
plt.ylabel("Y-axis")
plt.show()

#%%

# question 03

# library to plot the graph
import matplotlib.pyplot as plt

# function to calculate the plot coordinates for each month
def plotValues(monthlyRepayment, annualRate, debtAmount): 
    monthly_rate = annualRate/12
    
    # debtBalance to store the remaining balance
    debtBalance = debtAmount
    interest = 0
    monthCount = 0
    
    # to store the list of the remaining debtBalance each month
    debtBalanceList = []
    debtBalanceList.append(debtBalance)
    monthlist = []
    monthlist.append(monthCount)
    
    # loop until all debtBalance is paid and assuming the infinite at the monthCount of 1000
    while(debtBalance > 0 and monthCount < 1000): 
        # calculate the months interest to pay
        interest = debtBalance * monthly_rate / 100
        # deduct the amount paid that after deducting the interest amount first
        debtBalance = debtBalance - (monthlyRepayment - interest)
        monthCount = monthCount + 1
        # keep appending the coordinates for the plot
        debtBalanceList.append(debtBalance)
        monthlist.append(monthCount)
    
    # if the monthCount is infinite, (Infinite assumed at 1000)
    if(monthCount == 1000):
      # strMsg for part (b) 
      strMsg = "Debt will never be paid"
    else: 
      strMsg = f"Repayment in {int(monthCount/12)} year and {monthCount%12}"
    # return the msg, monthlist and debtBalanceList for plotting on the graph
    return strMsg, monthlist, debtBalanceList

def plotGraph(monthlyPayment, colors): 
  # figure size of graph to be 8 by 6
  fig, ax = plt.subplots(figsize = (8, 6))
  # define x and y axis range to be shown
  plt.xlim([0, 60])
  plt.ylim([0, 100000])
  # loop through the passed monthlyPayments to plot for
  i = 0
  while(i < len(monthlyPayment)): 
    monthlyPay = monthlyPayment[i]
    colorStr = colors[i]
    # call the plotValues() function to return the desired coordinates corresponding to the monthlyPay
    strMsg, x, y = plotValues(monthlyPay, 6.5, 100000)
    # plot line graph with the give color
    ax.plot(x, y, color = colorStr, label = 'R'+str(monthlyPay))
    i=i+1
  plt.xlabel('number of months')
  plt.ylabel('Debt')
  leg = plt.legend(loc = 'lower left')
  plt.show()

def main(): 
  monthlyPayment = [1000,2000,3000,4000]
  colors = ['blue', 'orange', 'green', 'brown']
  print("Part (a)")
  plotGraph(monthlyPayment, colors)
  print("\nPart (b)")
  i = 0
  while(i<len(monthlyPayment)): 
    strMsg, x, y = plotValues(monthlyPayment[i], 6.5, 100000)
    print(f"R{monthlyPayment[i]}: {strMsg}")
    i = i+1
  
  # testing part (3)
  print("\nPart (c)")
  monthRepay = 100
  strMsg, x, y = plotValues(monthRepay, 6.5, 100000)
  print(f"R{monthRepay}: {strMsg}")

```
